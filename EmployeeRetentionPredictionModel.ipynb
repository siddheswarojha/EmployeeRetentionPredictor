```python
#importing necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as nmp
from sklearn import linear_model
from sklearn.model_selection import train_test_split  
```

```python
df = pd.read_csv("HR_comma_sep.csv")
df
```

```python
df.head()
```

```python
df.tail()
```

```python
#adding dummy data to handle categorical values
newDF= pd.get_dummies(df.salary)
newDF
```

```python
#merging the both dataframes
finalDF = pd.concat([df,newDF],axis='columns')
finalDF
```

```python
bestDF = finalDF.drop(['salary','low'],axis='columns')
bestDF
```

```python
X= bestDF.drop(['left','Department'],axis='columns')
X
```

```python
y=bestDF.left
y
```

```python
#splitting the data for training and testing 
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
model = linear_model.LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
#accuracy
model.score(X_test,y_test)
```
